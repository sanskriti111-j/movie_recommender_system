```python
import ast
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movie['genres']=movie['genres'].apply(convert)
```

```python
movie.head()
```

```python
movie['keywords']=movie['keywords'].apply(convert)
```

```python
movie.head()
```

```python
import ast
import pandas as pd

def convert1(obj):
    L = []
    try:
        cast_list = ast.literal_eval(obj)
        for i in cast_list[:3]:  # Only loop through the first 3 elements
            L.append(i.get('name', ''))
    except (ValueError, SyntaxError, TypeError):
        # Handles invalid strings or NaN values
        pass
    return L

# Apply the function to the 'cast' column
movie['cast'] = movie['cast'].apply(convert1)
```

```python
movie['cast']
```

```python
import ast
import pandas as pd

def fetch_director(obj):
  L=[]
  if obj is not None:  # Add this check for None values
    try:
      for i in ast.literal_eval(obj):
        if i['job']=='Director':
          L.append(i['name'])
          break
    except (ValueError, SyntaxError, TypeError):
        # Handles invalid strings or NaN values
        pass
  return L
```

```python
movie['crew']=movie['crew'].apply(fetch_director)
```

```python
movie['overview']=movie['overview'].apply(lambda x:x.split() if isinstance(x, str) else [])
```

```python
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","") for i in x] )
movie['keywords']=movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x] )
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","") for i in x] )
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","") for i in x] )
```

```python
movie.head()
```

```python
movie['tag']=movie['overview']+movie['genres']+movie['keywords']+movie['cast']+movie['crew']
```

```python
movie.head()
```

```python
df=movie[['id','title','tag']]
```

```python
df['tag']=df['tag'].apply(lambda x:" ".join (x))
```

```python
df
```

```python
df['tag']=df['tag'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(df['tag']).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
df['tag']=df['tag'].apply(stem)
```

we find diatance between movies how much they are similar to each other. we willl not use eucledian distance instead we will use cosine distance as eucledian distance is not good when there are more dimensions

diatance inversely  prop to similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity[0]
```

```python
def recommend(movie):
  movie_index=df[df['title']==movie].index[0]
  distances=similarity[movie_index]
  movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movie_list:
    print(df.iloc[i[0]].title)
```

```python
recommend('Spectre')
```

```python
recommend('John Carter')
```
